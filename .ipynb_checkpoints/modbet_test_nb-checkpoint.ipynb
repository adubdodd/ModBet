```python
def draftkings_parse(input_dict, page, key):
    outcomes = [x.text for x in page.findAll(input_dict['outcomes_obj'],input_dict['outcomes_class'])]
    odds = [x.text for x in page.findAll(input_dict['odds_obj'],input_dict['odds_class'])]
    return list(zip(outcomes[::3], outcomes[2::3], odds[::3], odds[1::3], odds[2::3], [key] * len(home_teams), [pd.datetime.now()] * len(home_teams)))

def fanduel_parse(input_dict, page, key):
    outcomes = [x.text for x in page.findAll(input_dict['outcomes_obj'],input_dict['outcomes_class'])]
    odds = [x.text for x in page.findAll(input_dict['odds_obj'],input_dict['odds_class'])]
    return list(zip(outcomes[::2], outcomes[1::2], odds[::3], odds[1::3], odds[2::3], [key] * len(home_teams), [pd.datetime.now()] * len(home_teams)))
```
